```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('./train.csv')
test  = pd.read_csv('./test.csv')
sub   = pd.read_csv('./submit.csv')
```

```python
train.drop('SEQ', axis=1, inplace=True)
test.drop('SEQ', axis=1, inplace=True)
```

```python
train['target'] = np.log(train['target'])
train.head()
```

```python
test.head()
```

```python
from sklearn.preprocessing import RobustScaler

rs = RobustScaler()
rs.fit(train[test.columns])
train[test.columns] = rs.transform(train[test.columns])
test[test.columns]  = rs.transform(test[test.columns])
```

```python
train.head()
```

```python
test.head()
```

```python
X = train.drop('target', axis=1)
y = train['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, shuffle=True, random_state=0)
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
from xgboost import XGBRFRegressor

model = XGBRFRegressor()
model.fit(X_train, y_train)
pred = model.predict(X_val)
print(pred.shape)
pred[:10]
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_val, pred, squared=False)
```

```python
from sklearn import metrics
metrics.SCORERS.keys()
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [100, 200]
}

gscv = GridSearchCV(
    model, params, 
    scoring='neg_root_mean_squared_error', 
    refit=True, cv=5
)

gscv.fit(X, y)
```
