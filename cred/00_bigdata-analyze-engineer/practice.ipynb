```python
boston_data = datasets.load_boston()
boston = pd.DataFrame(boston_data['data'], columns=boston_data['feature_names'])
boston['target'] = boston_data['target']
boston
```

```python
upper = boston['target'].mean() + 1.5 * boston['target'].std()
lower = boston['target'].mean() - 1.5 * boston['target'].std()

print(boston[(boston['target'] < upper) & (boston['target'] > lower)]['AGE'].std())
```
