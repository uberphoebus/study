# 8. 막대 그래프(기본)

```python
lab = ['강백호', '서태웅', '정대만']
val = [190, 187, 184]
ticks = ['1번', '2번', '3번']

plt.bar(lab, val, color=['r', 'g', 'b'], alpha=0.5, width=0.5)
plt.ylim(180, 195)
plt.xticks(lab, ticks, rotation=45)
plt.yticks(rotation=45)

plt.show()
```

# 9. 막대 그래프(심화)

```python
plt.barh(lab, val)
```

```python
bar = plt.bar(lab, val) # 막대 객체

bar[0].set_hatch('/') # 패턴 채우기
bar[1].set_hatch('x')
bar[2].set_hatch('..')

for idx, rect in enumerate(bar):
    print(rect)
    plt.text(idx, rect.get_height() + 0.5, val[idx], ha='center', color='blue')

plt.ylim(175, 190)

plt.show()
```

# 13. 원 그래프(기본)

```python
val = [30, 25, 20, 13, 10, 2]
lab = ['python', 'java', 'js', 'c#', 'c/c++', 'etc']
explode = [0.2, 0.1, 0, 0, 0, 0]

plt.pie(
    val, labels=lab, autopct='%.1f%%', 
    startangle=90, counterclock=False, explode=explode
)
plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```

# 14. 원 그래프(심화)

```python
def custom_ap(pct):
    return ('%.1f%%' % pct) if pct >= 10 else ''

colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
wedgeprops = {'width':0.6, 'edgecolor':'w', 'linewidth':2}
plt.pie(
    val, labels=lab, autopct=custom_ap, 
    startangle=90, counterclock=False, 
    explode=explode, colors=colors, wedgeprops=wedgeprops,
    pctdistance=0.7
)
plt.legend(loc=(1.2, 0.3), title='언어')
plt.show()
```

# 17. Matplotlib 퀴즈

```python
quiz = pd.DataFrame(
    {
        '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
        '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
        '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
        '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
    }
)
```

```python
plt.title('국내 Top 8 영화 평점 정보')
plt.bar(quiz['영화'], quiz['평점'])
plt.xticks(rotation=90)
plt.xlabel('영화')
plt.ylabel('평점')
```

```python
year = quiz.groupby('개봉 연도')['평점'].mean().index
val = quiz.groupby('개봉 연도')['평점'].mean().values
plt.plot(year, val, marker='o')
plt.xticks([2005, 2010, 2015, 2020])
plt.ylim(7, 10)
plt.show()
```

```python
filt = quiz['평점'] >= 9.0
val = [len(quiz[filt]), len(quiz[~filt])]
lab = ['9점 이상', '9점 미만']

plt.pie(val, labels=lab, autopct='%.1f%%')
plt.legend(loc=(1, 0.3))
```
