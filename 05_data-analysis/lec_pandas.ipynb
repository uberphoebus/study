# 13. 함수 적용

```python
temp = df.copy()

def add_cm(height):
    return str(height) + 'cm'

temp['키'] = temp['키'].apply(add_cm)
temp
```

```python
def cap(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

temp['SW특기'] = temp['SW특기'].apply(cap)
temp
```

# 15. 퀴즈

```python
quiz = pd.DataFrame(
    {
        '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
        '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
        '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
        '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
    }
)
```

```python
quiz[['영화', '평점']]
```

```python
quiz.loc[quiz['개봉 연도'] >= 2015, ['영화', '개봉 연도']]
```

```python
quiz['추천 점수'] = quiz['관객 수'] * quiz['평점'] // 100
quiz
```

```python
quiz.sort_values('개봉 연도', ascending=False)
```
