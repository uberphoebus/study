# Basic Plot

```python
bp = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000) )  
```

```python
bp = bp.cumsum()
```

```python
bp.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000,10), index=bp.index, columns=list("ABCDEFGHIJ"))
df = df.cumsum()
```

```python
plt.figure();
df.plot();
```

```python
df3 = pd.DataFrame(np.random.randn(1000,2), columns=["B","C"]).cumsum()
df3["A"] = pd.Series(list(range(len(df))))
df3.plot(x="A", y="B");
```

# Scatter Plot

# Hexagonal bin plot

```python
hp = pd.DataFrame(np.random.rand(1000,2), columns =["a", "b"])
hp["b"] = hp["b"] + np.arange(1000)
hp.plot.hexbin(x ="a", y="b", gridsize=25)
```

```python
hp2 = pd.DataFrame(np.random.randn(1000,2), columns=["a","b"])
hp2["b"] = hp2["b"] + np.arange(1000)
hp2["z"] = np.random.uniform(0,3,1000)
hp2.plot.hexbin(x = "a", y= "b", C="z", reduce_C_function=np.max, gridsize=25)
```

# Lag plot

```python
from pandas.plotting import lag_plot
plt.figure();
spacing = np.linspace(-99 * np.pi, 99 * np.pi, num=1000)
data = pd.Series(0.1 * np.random.rand(1000) + 0.9 * np.sin(spacing))
lag_plot(data);
```

# Autocorrelation plot

```python
from pandas.plotting import autocorrelation_plot
plt.figure();
spacing = np.linspace(-9 * np.pi, 9 * np.pi, num=1000)
data = pd.Series(0.7 * np.random.rand(1000) + 0.3 * np.sin(spacing))
autocorrelation_plot(data);
```

# Bootstrap plot

```python
from pandas.plotting import bootstrap_plot
data = pd.Series(np.random.rand(1000))
bootstrap_plot(data, size=50, samples=500, color="blue")
```
