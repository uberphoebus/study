### ( 실습.. DataFrame 만들기)
1. 날짜, 사망자수, 확진자수, 지역발생수, 해외유입수
2. createDt, deathCnt, defCnt, localOccCnt, overFlowCnt

```python
import pandas as pd
import requests 
import xmltodict 
import json 

end_point = "http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19SidoInfStateJson?serviceKey="
params = "&pageNo=1&numOfRows=10&startCreateDt=20210801&endCreateDt=20210801"

url = end_point + my_key + params 

res = requests.get(url)
result = xmltodict.parse(res.text)
json_dict = json.loads(json.dumps(result))
list_data = json_dict["response"]["body"]["items"]["item"] 

df = pd.DataFrame(list_data)
df1 = df[["createDt","deathCnt","defCnt","localOccCnt","overFlowCnt"]]
df1.columns = ["날짜","사망자수","확진자수"," 지역발생수","해외유입수"]
df1.head()
```
