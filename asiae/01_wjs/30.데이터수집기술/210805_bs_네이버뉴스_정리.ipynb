```python
## 네이버 뉴스 속보를 수집 
import requests 
from bs4 import BeautifulSoup 
headers = { "User-Agent":"Mozilla/5.0"}

def get_page(url):    
    html = requests.get(url, headers=headers)
    soup = BeautifulSoup(html.text)  # "html.parser"  
    type06 = soup.find("ul",{"class":"type06_headline"})
    dl = type06.find_all("dl")
    return dl

def page_to_items(dl):
    news_page = []

    for item2 in dl:
        try:
            # 기사 이미지 
            img = item2.find("dt",{"class":"photo"}).find("img")
            img_src = img["src"]      
            # 기사 링크, 타이틀 
            link = item2.find("dt", {"class":""}).find("a")
            link_href = link["href"]        
            link_text =link.text.replace("\t","").replace("\r\n","").strip()     
            # 기사 내용 
            content = item2.find("dd")
            content_text = content.text.replace("\n","").split("…")[0] 

            news_page.append([img_src, link_href, link_text, content_text])

        except: 
            print("No Image or No Contents!! ")
    return news_page
```

```python
import pandas as pd
import time 

base_url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"
date = ["20210803","20210804"]
page = [1,2,3]

init = 0
for i in date:
    for j in page:
        url = base_url + "&date=" + i + "&page=" + str(j)
#        print(url)
        
        time.sleep(0.5)
        get_dl = get_page(url)
        news_page = page_to_items(get_dl)        

        news_page_df = pd.DataFrame(news_page, 
                      columns = ["이미지링크","기사링크","기사제목","기사내용"])
        news_page_df["날짜"] = i
        news_page_df["page"] = j 
        
        if init ==0 : 
            total_df = news_page_df.copy()
            init = 1
        else:
            total_df = pd.concat([total_df, news_page_df], ignore_index=True)
        
        print("page_df = ", news_page_df.shape)
        print("total_df = ", total_df.shape)
```

```python
import os 
def createfolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print("Error : Creating directory. " + directory) 
        
import urllib.request 
createfolder("c:/jswoo/img_down/")      

for row in range(0, len(total_df)):
    img_file = total_df.iloc[row,4].split("?")
    img_name = img_file[0].split("/")[-1]
    down_filename = "c:/jswoo/img_down/" + img_name 
    urllib.request.urlretrieve(img_file[0], down_filename)
    if row%10==0:
        print("down = ", img_file[0], " ** ", down_filename)
        
print("완료!!! ")
```
