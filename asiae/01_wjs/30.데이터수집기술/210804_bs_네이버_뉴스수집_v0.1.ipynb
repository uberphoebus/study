```python
## 네이버 뉴스 속보를 수집 
import requests 
from bs4 import BeautifulSoup 
headers = { "User-Agent":"Mozilla/5.0"}
url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"

# 해당 url requests.get 
html = requests.get(url, headers=headers)

soup = BeautifulSoup(html.text)  # "html.parser"  
type06 = soup.find("ul",{"class":"type06_headline"})
dl = type06.find_all("dl")
```

```python
# 기사 이미지 주소 수집 
for item2 in dl:
    try:
        img = item2.find("dt",{"class":"photo"}).find("img")
        print("img = ", img["src"])
    except:
        print(" no Image ")
```

```python
# 기사 링크 주소 및 제목 
for item2 in dl:
    try:
        link = item2.find("dt", {"class":""}).find("a")
        print(link["href"])
        print(link.text.replace("\t","").replace("\r\n","").strip())
    except:
        print("No link or Title !!")
```

```python
# 기사 본문 내용 
item2 = dl[0]

for item2 in dl:
    try:
        content = item2.find("dd")
        print(content.text.replace("\n","").split("…")[0])
    except:
        print(" No Content !!")
```
