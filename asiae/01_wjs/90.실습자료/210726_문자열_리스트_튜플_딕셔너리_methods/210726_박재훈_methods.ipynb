# 02 List methods

```python
fru = ['app', 'ban', 'cher']
fru.append('oran') # 값 추가
fru
```

```python
fru.clear() # 리스트 비우기
fru
```

```python
fru = ['app', 'ban', 'cher', 'oran']
x = fru.copy() # 리스트 복사
x
```

```python
fru.count('ban') # 값 개수 조회
```

```python
car = ['bmw', 'hd', 'kia']
fru.extend(car) # 리스트1에 리스트2 병합
fru
```

```python
fru.index('bmw') # 문자열 인덱스 출력
```

```python
fru.insert(4, 'grap') # n번 인덱스에 값 삽입
fru
```

```python
fru.pop(4) # n번 인덱스 삭제
fru
```

```python
fru.remove('bmw') # 값 삭제
fru
```

```python
fru.reverse() # 역순 정렬
fru
```

```python
fru.sort() # 정렬
fru
```

```python
fru.sort(reverse=True) # 역순 정렬
fru
```

# 03 Dictionary methods

```python
car = {'brnd': 'hdy', 'mod': 'ioniq', 'year': '2021'}
car.clear() # 딕셔너리 요소 삭제
car
```

```python
car = {'brnd': 'hdy', 'mod': 'ioniq', 'year': '2021'}
x = car.copy() # 딕셔너리 복사
x
```

```python
x = ('k1', 'k2', 'k3')
y = 3
xy = dict.fromkeys(x, y) # 키와 값 변수로 딕셔너리 구성
xy
```

```python
car.get('mod') # 키로 값 검색
```

```python
car.items() # 키-값 출력
```

```python
car.keys() # 키만 출력
```

```python
car.values() # 값만 출력
```

```python
car.pop('mod') # 키로 아이템(키-값) 삭제
car
```

```python
car.popitem() # 마지막 아이템 삭제
car
```

```python
car = {'brnd': 'hdy', 'mod': 'ioniq', 'year': '2021'}
car.setdefault('month', '5')
# 특정 키의 값을 특정하게 수정
# 키가 존재하지 않으면 아이템 삽입
car
```

```python
car.update({'col': 'blue'}) # 아이템 삽입
car
```
