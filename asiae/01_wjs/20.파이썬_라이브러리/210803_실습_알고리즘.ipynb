```python
a = [2,4,5,1,3]
    
for i in range(0, len(a)-1):
    print("i = ", a[i]) 
    min_idx = i
    for j in range(i+1, len(a)):
        if a[j] < a[min_idx]:
            min_idx = j        
        print("i = {}, j = {}, min_idx = {}".format(a[i],a[j], a[min_idx]))
    a[i], a[min_idx] = a[min_idx], a[i]
    print("a = ", a)
    print("-"*20)
```

### ( 실습. sel_sort() 함수를 만들어서 실행)

```python
def sel_sort(a):
    for i in range(0, len(a)-1):
        min_idx = i
        for j in range(i+1, len(a)):
            if a[j] < a[min_idx]:
                min_idx = j 
        a[i], a[min_idx] = a[min_idx], a[i]
    print("sort result = ", a)
    return a
```

```python
a = [2,4,5,1,3]
sel_sort(a)
```

```python
b = sel_sort(a)
print("sort 후 = ", b)
```
