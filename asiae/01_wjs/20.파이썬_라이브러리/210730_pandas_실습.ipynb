## ( 실습 성적처리 ) 

```python
# df 만들기 
import pandas as pd

score_dic = {"국어":[76,90,64,70,98],
            "영어":[65,75,70,82,93],
            "수학":[80,98,62,84,85]}
df = pd.DataFrame(score_dic)
df
```

```python
data = [[76,65,80],
        [90,75,98],
        [64,70,62],
        [70,82,84],
        [98,93,85]]
df1 = pd.DataFrame(data, columns=["국어","영어","수학"])
df1
```

```python
# 총점 구하기 
df["총점"] = df["국어"] + df["영어"] + df["수학"]
df
```

```python
# 평균 구하기 
df["평균"] = (df["국어"] + df["영어"] + df["수학"])/3
df
```

```python
df["평균1"] = df.iloc[:,:3].mean(axis=1)
df
```

```python
# 평균을 등급으로 
df["등급"] = ""
df
```

```python
df["등급"] = ""
for row in range(0, len(df)):
    #print(df.iloc[row,4])
    if df.iloc[row,4] > 90:
        df.iloc[row,-1] = "S등급"
    elif df.iloc[row,4] > 80:
        df.iloc[row,-1] = "A등급"
    elif df.iloc[row,4] > 70:
        df.iloc[row,-1] = "B등급"
    else:
        df.iloc[row,-1] = "C등급"
df
```

```python
df["등급1"] = ""
for row in range(0, len(df)):
    #print(df.iloc[row,4])
    in_pos = df.iloc[row,4]
    if in_pos > 90:
        df.iloc[row,-1] = "S등급"
    elif in_pos > 80:
        df.iloc[row,-1] = "A등급"
    elif in_pos > 70:
        df.iloc[row,-1] = "B등급"
    else:
        df.iloc[row,-1] = "C등급"
df
```

```python
df.loc[df["평균"] > 90, "등급2"] = "S등급"
df.loc[(df["평균"] > 80) & (df["평균"] <=90), "등급2"] = "A등급" 
df.loc[(df["평균"] > 70) & (df["평균"] <=80), "등급2"] = "B등급" 
df.loc[df["평균"] <=70, "등급2"] = "C등급" 
df
```
