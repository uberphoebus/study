```python
n = '71'

if int(n) < 10:
    n = n + '0'

i = 0
start = n

while True:
    i += 1
    a, b = n[0], n[-1]
    c = int(a) + int(b)
    n = b + str(c)[-1]
    
    if n == start:
        print(i)
        break
```

```python
input = '7 100 95 90 80 70 60 50'

temp = [int(i) for i in input.split()]
m = sum(temp[1:]) / temp[0]

res = round([j > m for j in temp].count(True) / temp[0] * 100, 3)

print(f'{res}%')
```

```python
sent = ['5 50 50 70 80 100', 
'7 100 95 90 80 70 60 50', 
'3 70 90 80', 
'3 70 90 81', 
'9 100 99 98 97 96 95 94 93 91', ]

for i in sent:
    temp = list(map(int, i.split()))
    m = sum(temp[1:]) / temp[0]
    res = [k > m for k in temp].count(True) / temp[0] * 100
    print('{:.3f}%'.format(res))
```

```python
a = 75

def d(a, n):
    temp = [a]
    for i in range(len(str(a))):
        temp.append(str(a)[i])
    result = sum(list(map(int, temp)))
    
    if result > n:
        return
    
    print(result)
    d(result, n)

d(20, 10000)
```
