```python
# pip install selenium
# conda install selenium
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys

driver = webdriver.Chrome('C:/Users/toto/Google Drive/sync/python/chromedriver')
url = 'https://www.naver.com'
driver.get(url)

assert 'NAVER' in driver.title # 타이틀에 해당 문자열이 없으면 오류

elem = driver.find_element_by_name('query')
elem.send_keys('python')
elem.send_keys(Keys.RETURN)

driver.page_source # 페이지소스 조회
# assert 'NO' not in driver.page_source
```

```python
# 브라우저 없이 실행
options = webdriver.ChromeOptions()
options.add_argument('headless')

driver = webdriver.Chrome('C:/Users/toto/Google Drive/sync/python/chromedriver', options=options)

driver.get('http://python.org')

res = driver.page_source

print(len(res))
```
