```python
# 실습01 sort 알고리즘(for 2번, if, range)
# 1. input -> [2, 4, 5, 1, 3]
# 2. output -> [1, 2, 3, 4, 5]
# 3. 순서대로 비교해서 크기 순으로 정렬되게 이동하는 알고리즘; sort 알고리즘

x = [2, 4, 5, 1, 3]

def comp(i, j):
    if x[i] > x[j]:
        temp = x.copy()
        temp[i:j+1]
        x.remove[i:j+1]

for i in range(0, 4):
    for j in range(1, 5):
        temp = x[j]
        if x[i] > x[j]:
            x.insert(i, temp)
            x.pop(j)
            print(x)
        else:
            continue
```

```python
# 실습02; sel_sort() 함수를 만들어서 실행

a = [2, 4, 5, 1, 3]

def sel_sort(a):
    for i in range(0, len(a)-1):
        min_idx = i
        for j in range(i+1, len(a)):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a
        
b = sel_sort(a)
print('sort 후 = ', b)
```

```python
# mandelbrot set ??
def mandelbrot(h, w, maxit=20, r=2):
    '''Returns an image of the Mandelbrot fractal of size (h,w)'''
    x = np.linspace(-2.5, 1.5, 4*h+1)
    y = np.linspace(-1.5, 1.5, 3*w+1)
    A, B = np.meshgrid(x, y)
    C = A + B*1j
    z = np.zeros_like(C)
    divtime = maxit + np.zeros(z.shape, dtype=int)
    
    for i in range(maxit):
        z = z**2 + C
        diverge = abs(z) > r
        div_now = diverge & (divtime == maxit)
        divtime[div_now] = i
        z[diverge] = r
    return divtime
plt.imshow(mandelbrot(400, 400))
```
