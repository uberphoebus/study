## Group 함수

<pre>
count(1)  --- value_counts()
avg()     --- mean()

max()     --- max()
min()     --- min()
sum()     --- sum()
</pre>

* select deptno,  avg(sal) from emp group by deptno;

```python
df.groupby(by="deptno") ["sal"].mean()
```

* select deptno,  min(sal) , max(sal) from emp group by deptno;

```python
# df.groupby(by="deptno") [    ['sal'].min() , ['comm'].max()  ] 
```

```python
df.groupby(by="deptno")["sal"].agg([max, min]) 
```

```python
dict = {"sal":"max",  "comm":"min"}
# dict = {"sal":   ['min', 'max']}
df.groupby(by="deptno").agg(dict) 
```

```python
df[ df['deptno']==10  ].groupby(by="deptno").agg(dict) 
```
