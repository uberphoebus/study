```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf= TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['content'])
print(tfidf_matrix.shape)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(cosine_sim)
```

```python
def getRecommend(title, cosine_sim=cosine_sim):
    
    # idx = title in title_srs['title']
    # idx = df['title'].str.contains(title)
    
    idx = df[df['title'].str.contains(title)].index.tolist()[0]
    
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    
    new_idx = [i[0] for i in sim_scores]
    scores  = [j[1] for j in sim_scores]
    
    df_rec = df.iloc[new_idx].drop('content', axis=1).copy()
    df_rec.reset_index()
    df_rec['cos_score'] = scores
    
    return df_rec
```

```python
def getTopwords(cat=1):
    
    list = []
    
    top_df = df[df['cate'] == cat]
    top_df = top_df.reset_index(drop=True)
    
    for i in range(top_df.shape[0]):
        
        # print(top_df['nouns'][i])
        
        for j in range(len(top_df['nouns'][i])):
            
            list.append(top_df['nouns'][i][j])
    
    top_words = pd.DataFrame({'word':list}).value_counts().head(5)
    
    return top_words
```

```python
for i in range(1, 7):
    print(getTopwords(i))
```
