# Usage Guide



figure; 시각화 영역

axes; 각 부분 영역

x/y axis label; 각 축 라벨

시각화 방법 두 가지
   - OO(object-oriented) style
   - pyplot style

https://matplotlib.org/stable/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py

```python
# a simple example
fig, ax = plt.subplots()  # 단일 ax fig 구성
x = [1, 2, 3, 4]; y = [1, 4, 2, 3]

ax.plot(x, y)  # .plot(*args)
plt.show()
```

```python
# oo style example

plt.rc('font', family='Malgun Gothic') # 한글 폰트 설정

# 데이터 준비
x = np.linspace(0, 2, 100) # 100개 간격으로 0 ~ 2 사이 난수 구성
print(x)

# 1. 시각화 영역 설정; 
fig, ax = plt.subplots() # 클래스 불러오기

# 2. 시각화; 여러 데이터 한 영역에 표현 가능
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

# 3. 시각화 설정
ax.set_xlabel('x축')  # Add an x-label to the axes.
ax.set_ylabel('y축')  # Add a y-label to the axes.
ax.set_title("간단한 그래프")  # Add a title to the axes.
ax.legend(loc=0)  # 라벨 표시, loc 0~10 설정 가능 
```

```python
# pyplot example

# 데이터 준비
x = np.linspace(0, 2, 100)

# 1. 시각화 영역 설정; pyplot은 영역 설정 불필요

# 2. 시각화
plt.plot(x, x, label='linear')  
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# 3. 시각화 설정
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend() # loc 설정 가능
```

```python
# 함수 시각화

def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out

# 데이터 준비
data1, data2, data3, data4 = np.random.randn(4, 100)

fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})

fig, (ax1, ax2) = plt.subplots(1, 2) # 1행 2열, 2개 ax 구성, s 주의
my_plotter(ax1, data1, data2, {'marker': 'x', 'color':'red'})
my_plotter(ax2, data3, data4, {'marker': 'o', # 지점 모양
                               'color':'green', # 색
                               'linewidth':5}) # 선 굵기
```

```python
# 한 figur에 그리기

import numpy as np

t = np.arange(0., 5., 0.2); t # 0~5까지 0.2간격으로 생성
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')
plt.plot(t, t**3, 'g^')
```

```python
# 한 줄에 작성
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
```

```python
# plotting with keyword strings

data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', # x
            'b', # y
            c='c', # 색상
            s='d', # 마커 크기
            alpha=0.5, # 농도
            data=data) # 원본 데이터
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.colorbar() # 색 인덱스 추가
plt.show()

# api 메뉴얼
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html
```

```python
# plotting with categorical varibales
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

# 서브플롯 정의 방법; 1행3열1번째
plt.subplot(231)
plt.bar(names, values)
plt.subplot(233)
plt.scatter(names, values)
plt.subplot(235)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# working with text; histogram
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, # 값
                            50, # bin, x축 간격
                            density=1, 
                            facecolor='g', # 색상
                            alpha=0.75) # 투명도


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, # position
         r'$\mu=100,\ \sigma=15$') # 설명
plt.axis([40, 160, 0, 0.03]) # 축 범위
plt.grid(True) # 격자 표시
plt.show()
```

```python
# piechart

# https://matplotlib.org/stable/tutorials/introductory/sample_plots.html#sphx-glr-tutorials-introductory-sample-plots-py


import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs' # 구분자
sizes = [15, 30, 45, 10] # 데이터의 크기
explode = (0.3, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

# fig1, ax1 = plt.subplots()
plt.pie(sizes, # 값 
        explode=explode, # 4개 분류에 차례대로 잘라지는 형태 부여
        labels=labels, # 분류, 구분, class
        autopct='%1.1f%%', # 출력 옵션
        shadow=True, # 그림자
        startangle=90 # 첫 값의 시작 각도
       )
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
index = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.bar(x, # x축
        values, # y측
        width=0.6, # 막대 폭
        align='edge', 
        color='springgreen', # 색상 테마
        edgecolor='gray', # 테두리 색상 테마
        linewidth=3, # 테두리 두께
        tick_label=index, # x축 라벨
        log=True) # 로그 형태
plt.xticks(x, index)
plt.show()
```

```python
# stacked barchart

x1 = [1, 2, 2, 1, 2]
x2 = [2, 3, 4, 2, 1]
plt.bar(range(len(x1)), x1)
plt.bar(range(len(x2)), x2, bottom=x1)
plt.legend(['x1', 'x2'])
plt.show()
```

```python
# titanic 성별 생존 분포 데이터
# 1. 생존자의 남여 분포; female, male 인원수
# 2. 사망자의 남여 분포; female, male 인원수

import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 1. 생존자로 구성된 데이터 구성
titanic[titanic.survived == 1]
```

```python
surv = titanic[titanic.survived == 1].sex.value_counts()
dead = titanic[titanic.survived == 0].sex.value_counts()
print(surv)
dead
```

```python
# 2. 구성된 데이터 합치기
import pandas as pd
pd.DataFrame([surv, dead])
```

```python
# 3. 합친 데이터 인덱스 부여
df = pd.DataFrame([surv, dead], 
                  index=['survived', 'dead'] # 인덱스 지정 문법 주의
                 )
```

```python
# 4. barchart 시각화
df.plot(kind='bar')
```

```python
# 5. stacked barchart 시각화
df.plot(kind='bar', stacked=True)
```

```python
# 6. 통합 정리
surv = titanic[titanic.survived == 0]['sex'].value_counts()
dead = titanic[titanic.survived == 1]['sex'].value_counts()

df = pd.DataFrame([surv, dead], index=['survived', 'dead'])
df.plot(kind='bar', stacked=True)
```

```python
# 컬럼별 생존 유무에 따른 분포 함수 만들기; stacked_barchart()
def stacked_barchart(cl):
    surv = titanic[titanic.survived == 0][cl].value_counts()
    dead = titanic[titanic.survived == 1][cl].value_counts()
    df = pd.DataFrame([surv, dead], index=['survived', 'dead'])
    return df.plot(kind='bar', stacked=True)

stacked_barchart('sex')
```

```python
print(stacked_barchart('pclass'))
print(stacked_barchart('sibsp'))
print(stacked_barchart('parch'))
print(stacked_barchart('embarked'))
```
