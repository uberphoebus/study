## try except

```python
# 잘못된 코드가 있을 때 제어하는 방법
try:
    print(111)
    print(1/0)
    print('222')
except:
    print('에러 발생')

def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print('0으로 나눌 수 없음')
    except TypeError:
        print('type error')
    else:
        print('결과는 = ', result)
    finally:
        print('에러와 무관하게 실행')

divide(2, 4); divide('2', 1)
```

## 네이버 뉴스 수집 v0.7
함수 구성

1. get_page(); url -> dl
2. page_to_items(); dl -> news_page(lst)
3. news_df 구성
4. news_df.csv 저장

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def get_page(url):
    headers = {'User-Agent':'Mozilla/5.0'}
    html = requests.get(url, headers=headers).text
    soup = BeautifulSoup(html, 'html.parser')

    type06 = soup.find('ul', {'class':'type06_headline'})
    dl = type06.find_all('dl')
    return dl

def page_to_items(dl):
    news_page = []
    for item in dl:
        try:
            img = item.find('dt', {'class':'photo'}).find('img')
            img_src = img['src']

            link = item.find('dt', {'class':''}).find('a')
            link_href = link['href']

            title = link.text.replace('\t', '').replace('\r\n', '').strip()

            content = item.find('dd')
            content_text = content.text.replace('\n', '').split('…')[0].strip()

            news_page.append([img_src, link_href, title, content_text])       
        except:
            pass
    return news_page

news_page = page_to_items(get_page(url))

news_df = pd.DataFrame(news_page,
                       columns=['img_l', 'link_l', 'title_l', 'content_l'])
# news_df.to_csv('news_df_v0.7.csv')
# news_df
```

```python
# 1번 방법; 함수 미수정
import pandas as pd
import requests
from bs4 import BeautifulSoup


def get_page(url):
    headers = {'User-Agent': 'Mozilla/5.0'}
    html = requests.get(url, headers=headers).text
    soup = BeautifulSoup(html, 'html.parser')

    type06 = soup.find('ul', {'class': 'type06_headline'})
    dl = type06.find_all('dl')
    return dl


def page_to_items1(dl):
    news_page = []
    for item in dl:
        try:
            img = item.find('dt', {'class': 'photo'}).find('img')
            img_src = img['src']

            link = item.find('dt', {'class': ''}).find('a')
            link_href = link['href']

            title = link.text.replace('\t', '').replace('\r\n', '').strip()

            content = item.find('dd')
            content_text = content.text.replace('\n', '').split('…')[0].strip()

            news_page.append([img_src, link_href, title, content_text])
        except:
            pass
    return news_page


import time

base_url = 'https://news.naver.com/main/list.naver?mode=LSD&mid=sec&sid1=001&date='
date = ['20210803', '20210804']
page = ['1', '2', '3']
init = 0

for i in date:
    for j in page:
        url = base_url + i + '&page=' + j
        time.sleep(0.3)  # 0.3초 간격 조회
        get_dl = get_page(url)
        news_page = page_to_items1(get_dl)

        news_page_df = pd.DataFrame(
            news_page, columns=['img1', 'link1', 'title1', 'content1'])
        news_page_df['date1'] = i
        news_page_df['page1'] = j

        print(news_page_df.shape)

        if init == 0:
            news_df = news_page_df.copy()
            init = 1
        else:
            news_df = pd.concat([news_df, news_page_df], ignore_index=True)

news_df_v08 = news_df[[
    'date1', 'page1', 'title1', 'content1', 'img1', 'link1'
]]
news_df_v08

# news_df_v08.to_csv('news_df_v0.8.csv', encoding='utf-8-sig', index=False)
# pd.read_csv('news_df_v0.8.csv')
```

```python
# 2번 방법; 함수 수정

import pandas as pd
import requests
from bs4 import BeautifulSoup


def get_page(url):
    headers = {'User-Agent': 'Mozilla/5.0'}
    html = requests.get(url, headers=headers).text
    soup = BeautifulSoup(html, 'html.parser')

    type06 = soup.find('ul', {'class': 'type06_headline'})
    dl = type06.find_all('dl')
    return dl


def page_to_items2(dl, date):
    news_page = []
    for item in dl:
        try:
            img = item.find('dt', {'class': 'photo'}).find('img')
            img_src = img['src']

            link = item.find('dt', {'class': ''}).find('a')
            link_href = link['href']

            title = link.text.replace('\t', '').replace('\r\n', '').strip()

            content = item.find('dd')
            content_text = content.text.replace('\n', '').split('…')[0].strip()

            news_page.append([date, img_src, link_href, title, content_text])
        except:
            pass
    return news_page


import time

base_url = 'https://news.naver.com/main/list.naver?mode=LSD&mid=sec&sid1=001&date='
date = ['20210803', '20210804']
page = ['1', '2', '3']
news_page_merge = []

for i in date:
    for j in page:
        url = base_url + i + '&page=' + j
        time.sleep(0.3)  # 0.3초 간격 조회
        get_dl = get_page(url)
        news_page = page_to_items2(get_dl, i)
        news_page_merge += news_page
        print(len(news_page_merge))
print('완료', len(news_page_merge))

news_df = pd.DataFrame(
    news_page_merge, columns=['date2', 'img2', 'link2', 'title2', 'content2'])
news_df
# news_df_v09 = news_df_v09[['date2', 'title2', 'content2', 'img2', 'link2']]
# news_df_v09.to_csv('news_df_v0.9.csv', encoding='utf-8-sig', index=False)
```
