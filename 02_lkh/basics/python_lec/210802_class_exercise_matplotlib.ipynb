```python
# 실습01; 홀짝 문자열 카운트
# 1. input -> [1, 2, 3, 'a', 6, 7, 'b']
# 2. output -> 홀수(3개), 짝수(2개), 문자열(2개)

x = [1, 2, 3, 'a', 6, 7, 'b']

even = 0
odd = 0
string = 0

for i in x:
    if type(i) == str:
        string += 1
    elif i % 2 == 0:
        even += 1
    else:
        odd += 1
# print(f'홀수 {len(odd)}개, 짝수 {len(even)}개, 문자열 {len(string)}개') # f 문법

# 리스트 형성하여 리스트의 개수 세는 방법
x = [1, 2, 3, 'a', 6, 7, 'b']

even_list = []; odd_list = []; string_list = []
for i in x:
    if type(i) == str:
        string_list.append(i)
    elif i % 2 == 0:
        even_list.append(i)
    else:
        odd_list.append(i)
print('홀수 {}개, 짝수 {}개, 문자열 {}개'.format(len(odd_list), len(even_list), len(string_list)))
```

```python
# 홀짝 문자열 카운트 예시답안
input_list = [1, 2, 3, 'a', 6, 7, 'b']

count = [0, 0, 0] # 홀수, 짝수, 문자열 순서
for i in input_list:
    if type(i) == str:
        count[2] += 1
    elif i % 2 == 0:
        count[1] += 1
    else:
        count[0] += 1
print(f'홀수 {count[0]}개, 짝수 {count[1]}개, 문자열 {count[2]}개')
```

```python
# 실습02; 위 문제를 input_count() 함수를 만들어 처리
# 1. 함수를 만든다.
# 2. x = [ ]
# 3. y = input_count(x)

x = [1, 2, 3, 'a', 6, 7, 'b']

def input_count(x):
    count = [0, 0, 0]
    for i in x:
        if type(i) == str:
            count[2] += 1
        elif i % 2 == 0:
            count[1] += 1
        else:
            count[0] += 1
    return count

y = input_count(x)
print('홀수 {}개, 짝수 {}개, 문자열 {}개'.format(y[0], y[1], y[2]))
```

```python
# 실습03; 리스트가 아닌 데이터, 개수를 알 수 없는 데이터
# 1. 1, 2, 3, 'ab', 'e'와 같은 방식
# 2. 몇 개가 들어올지 알 수 없음
# 3. *args

def input_count1(*args):
    even = 0 ; odd = 0 ; string = 0
    for i in args:
        if type(i) == str:
            string += 1
        elif i % 2 == 0:
            even += 1
        else:
            odd += 1
    return (odd, even, string)

input_count1(1, 2, 3, 'ab', 'e')
```

```python
# 실습04; 윤년 계산
# 1. input = 2020
# 2. ountput -> ????년은 윤년입니다. / ????년은 평년입니다.
# 3. 계산방법; 4로 나눠서 떨어지고, 100으로 나누어 떨어지지 않으며, 연도가 400으로 나누어 떨어지는 연도

year = 2020

if year % 4 != 0:
    leap = '평년'
elif year % 100 == 0:
    leap = '평년'
elif year % 400 != 0:
    leap = '평년'
else:
    leap = '윤년'
print('{}년은 {}입니다.'.format(year, leap))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            res = '윤년'
        else:
            res = '평년'
    else:
        res = '윤년'
else:
    res = '평년'
print(res)

# 한번에 정리
if ((year % 4 == 0) & (year % 100 != 0)) | (year % 400 == 0):
    res = '윤년'
else:
    res = '평년'
print(res)
```

```python
# 함수 만들어 그리기
plt.rc('font', family='Malgun Gothic')

def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1, 1)
my_plotter(ax, data1, data2, {'marker': 'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
my_plotter(ax1, data1, data2, {'marker': 'x', 'color':'red'})
my_plotter(ax2, data3, data4, {'marker': 'o', 'color':'green'})
```

```python
# https://matplotlib.org/stable/tutorials/introductory/sample_plots.html#sphx-glr-tutorials-introductory-sample-plots-py
# 바 차트, 파이차트 자주 사용

import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs' # 구분자
sizes = [15, 30, 45, 10] # 데이터의 크기
explode = (0.3, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

# fig1, ax1 = plt.subplots()
plt.pie(sizes, # 값 
        explode=explode, # 4개 분류에 차례대로 잘라지는 형태 부여
        labels=labels, # 분류, 구분, class
        autopct='%1.1f%%', # 출력 옵션
        shadow=True, # 그림자
        startangle=90 # 첫 값의 시작 각도
       )
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
