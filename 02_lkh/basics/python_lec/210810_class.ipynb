## 실습
1. title 컬럼을 구성해서, 직책 데이터 추출
2. 반복문; 각 row의 Name 데이터를 찾아서, 샘플로 직책 추출 알고리즘 적용하여, title 컬럼에 넣기

```python
for i in range(len(titanic_df)):
    name = titanic_df.loc[i, 'Name']
    name_to_title = name.split(',')[1].split('.')[0].strip()
    titanic_df.loc[i, 'title'] = name_to_title
```

```python
titanic_df['title'].value_counts()
```

```python
# title01; 데이터 양이 많으면 반복문 활용 불가능
# 정규표현식; 문자열화 후 해당 패턴을 가진 부분 추출
# 한글자 이상과 . 으로 구성된 문자열
# for보다 df함수를 활용하는 것이 더 중요. 속도도 빠름
# expand=False 
titanic_df['title01'] = titanic_df['Name'].str.extract('([A-Za-z]+)\.', expand=False)
```

```python
titanic_df.head()
```

```python
titanic_df['title01'].value_counts()
```

```python
# title02; Mr(0), Miss(1), Mrs(2), 나머지는 (3) 할당
for i in range(len(titanic_df)):
    a = titanic_df.loc[i, 'title']
    b = titanic_df.loc[i, 'title02']
    if titanic_df.loc[i, 'title'] == 'Mr':
        titanic_df.loc[i, 'title02'] = 0
    elif titanic_df.loc[i, 'title'] == 'Miss':
        titanic_df.loc[i, 'title02'] = 1
    elif titanic_df.loc[i, 'title'] == 'Mrs':
        titanic_df.loc[i, 'title02'] = 2
    else:
        titanic_df.loc[i, 'title02'] = 3
titanic_df.head()
```

```python
# .map 활용; srs에만 적용. .apply .applymap 함수 참고
title_mapping = {'Mr':0, 'Miss':1, 'Mrs':2, 'Master':3, "Dr": 3, 
                 "Rev": 3, "Col": 3, "Major": 3, "Mlle": 3, "Countess": 3, 
                 "Ms": 3, "Lady": 3, "Jonkheer": 3,"Don": 3, "Mme": 3,
                 "Capt": 3,"Sir": 3 } # 모든 것에 대해

titanic_df['title05'] = titanic_df['title'].map(title_mapping)
titanic_df['title05'].value_counts()
# https://pandas.pydata.org/docs/reference/api/pandas.Series.map.html
# https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html#pandas.Series.apply
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html#pandas.DataFrame.apply
```

```python
# Sex 컬럼
# 1. sex01 컬럼 구성
# 2. male(0), felmale(1)

sex_mapping = {'male':0, 'female':1}
titanic_df['sex01'] = titanic_df['Sex'].map(sex_mapping)
titanic_df.head()
```

```python
# age 컬럼
# 1. 결측치가 있음
# 2. 수치형 데이터

titanic_df[titanic_df['Age'].isnull()].head()
```

```python
# title별로는 나이가 비슷할 것
# title별로 나이의 중앙값 확인
titanic_df['age01'] = titanic_df['Age']
titanic_df['age01'].fillna(titanic_df.groupby(by='title')['age01'].transform('median'), inplace=True)
titanic_df.head()
```

```python
# 함수 만들기 kde_chart()
# 인자는 컬럼명, start_age, end_age
# kde_chart('age01', 0, 20)

def kde_chart(column, start_age, end_age):
    facet = sns.FacetGrid(titanic_df, hue="Survived",aspect=4) # aspect 높이와 넓이의 비율
    facet.map(sns.kdeplot, column, shade= True)
    facet.set(xlim=(0, titanic_df[column].max()))
    facet.add_legend()
    plt.xlim(start_age, end_age) # x축 범위
```

```python
# 나이대를 어떻게 그룹화할 것인지
# age02 만들기
# 16 이하는 0, 16~26 이하는 1, 26~36 이하는 2, 36~62 이하는 3, 62 초과는 4
# child(0), young(1), adult(2), mid_age(3), senior(4)
# 중요중요

titanic_df.loc[titanic_df['age01'] <= 16, 'age02'] = 0
titanic_df.loc[(titanic_df['age01'] > 16) & (titanic_df['age01'] <= 26), 'age02'] = 1
titanic_df.loc[(titanic_df['age01'] > 26) & (titanic_df['age01'] <= 36), 'age02'] = 2
titanic_df.loc[(titanic_df['age01'] > 36) & (titanic_df['age01'] <= 62), 'age02'] = 3
titanic_df.loc[titanic_df['age01'] > 62, 'age02'] = 4

print(titanic_df['age02'].value_counts())
titanic_df.iloc[:, 4:].head()
```

```python
# embarked 처리
# 1. 결측치 처리.. 'S' 대체
# 2. S(0), C(1), Q(2) 변경, embarked02 컬럼 구성

titanic_df['embarked01'] = titanic_df.Embarked.fillna('S')

titanic_df.loc[titanic_df.Embarked == 'S', 'embarked02'] = 0
titanic_df.loc[titanic_df.Embarked == 'C', 'embarked02'] = 1
titanic_df.loc[titanic_df.Embarked == 'Q', 'embarked02'] = 2

# .map 함수도 가능; 실습해볼 것
titanic_df.iloc[:, 4:].head()
titanic_df.embarked02.value_counts()
```

```python
# fare 처리; 요금 계산
# 1. 결측치는 Pclass의 중앙값으로 대체; fare01
# 2. kde_chart(); 0~20, 0~30, 0~500
# 3. 17 이하 0, 17-30 1, 30-100 2, 100 이상 3; fare02

titanic_df['fare01'] = titanic_df['Fare']
titanic_df['fare01'].fillna(titanic_df.groupby(by='Pclass')['fare01'].transform('median'), inplace=True)

titanic_df.loc[titanic_df.Fare <= 17, 'fare02'] = 0
titanic_df.loc[(titanic_df.Fare > 17) & (titanic_df.Fare <= 30), 'fare02'] = 1
titanic_df.loc[(titanic_df.Fare > 30) & (titanic_df.Fare <= 100), 'fare02'] = 2
titanic_df.loc[titanic_df.Fare > 100, 'fare02'] = 3

titanic_df.head()
```

```python
# cabin
# 1. 첫 문자열 추출하여 cabin01
# 2. 매핑, cabin02 {'A':0, 'B':0.4, 'C':0.8, 'D':1.2, 'E':1.6, 'F':2, 'G':2.4, 'T', 2.8}
# 3. NaN은 Pclass의 cabin02 의 중앙값으로 결측치 채워서 cabin03

titanic_df['cabin01'] = titanic_df['Cabin'].str[0]

cabin_mapping = {'A':0, 'B':0.4, 'C':0.8, 'D':1.2, 'E':1.6, 'F':2, 'G':2.4, 'T':2.8}

titanic_df['cabin02'] = titanic_df['cabin01'].map(cabin_mapping)

titanic_df['cabin03'] = titanic_df['cabin02']
titanic_df['cabin03'].fillna(titanic_df.groupby(by='Pclass')['cabin02'].transform('median'), inplace=True)
```

```python
# Sibsp, Parch 컬럼
# 1. familysize = Sibsp + Parch + 1
# 2. mapping 코드 만들기; familysize01
# {1:0, 2:0.4, 3:0.8, 4:1.2, 5:1.6, 6:2, 7:2.4, 8:2.8, 9:3.2, 10:3.6, 11:4}

titanic_df['familysize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1

familysize_mapping = {1:0, 2:0.4, 3:0.8, 4:1.2, 5:1.6, 6:2, 7:2.4, 8:2.8, 9:3.2, 10:3.6, 11:4}

titanic_df['familysize01'] = titanic_df.familysize.map(familysize_mapping)
```

```python
# 필요 데이터 정리
# titanic_new_df; survived, pclass, title02, sex01, age02, fare02, embarked02, cabin03, familysiex01
# 컬럼이름 변경; 유지, 유지, title, sex, age, fare, embarked, cabin, faam
# csv만들기 titanic_new.csv
# 읽어서 확인

titanic_df_new = titanic_df[['Survived', 'Pclass', 'title02', 'sex01', 'age02', 'fare02', 'embarked02', 'cabin03', 'familysize01']]
columns = {'Survived':'Survived', 'Pclass':'Pclass', 'title02':'title', 'sex01':'sex', 'age02':'age', 'fare02':'fare', 'embarked02':'embarked', 'cabin03':'cabin', 'familysize01':'familysize'}
titanic_df_new.rename(columns=columns, inplace=True)

# titanic_df_new.columns = [] 로 컬럼명 설정

titanic_df_new.to_csv('titanic_new.csv', index=False)
titanic_df.to_csv('titanic_df.csv', index=False)

df = pd.read_csv('titanic_new.csv'); df
```
