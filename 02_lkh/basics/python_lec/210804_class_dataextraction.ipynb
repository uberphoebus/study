```python
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print('0으로 나눌 수 없음')
    except TypeError:
        print('type error')
    else:
        print('결과는 = ', result)
    finally: # 생략 가능
        print('에러와 무관하게 실행')

divide(2, 4)
divide('2', 1)
```

```python
# 실습; 네이버 뉴스 수집
# 1. 전체 모듈; 라이브러리 정의, 목표 url 설정, url 데이터 가져오고(requests.get()), bs 적용(html.parser) 형태로 만들고, type06 가져오기
# 2. for 3번 적용 -> 1번으로 변경 출력
# 3. 이미지 주소(img_src), 기사 링크(link_href),기사 제목(link_text), 기사 내용(content_text) 가져오기

import requests
from bs4 import BeautifulSoup
headers = {'User-Agent':'Mozilla/5.0'}
url = "https://news.naver.com/main/list.naver?mode=LSD&mid=sec&sid1=001"

html = requests.get(url, headers=headers).text
soup = BeautifulSoup(html, 'html.parser')
type06 = soup.find('ul', {'class':'type06_headline'})

dl = type06.find_all('dl')
for item2 in dl:
    try:
        img = item2.find('dt', {'class':'photo'}).find('img')
        img_src = img['src']
        print('이미지링크 = ', img_src)
        
        link = item2.find('dt', {'class':""}).find('a')
        link_href = link['href']
        print('기사링크 = ', link_href)
        
        link_text = link.text.replace('\t', '').replace('\n', '').split('…')[0]
        print('기사제목 = ', link_text)
        
        content_text = item2.find('dd').text.replace('\t', '').replace('\n', '').split('…')[0]
        print('기사본문 = ', content_text)
        print('-'*90)
    except:
        print('No Image')
```

# bs 네이버 뉴스 수집 함수로 만들기 v0.7 

```python
# 실습;
# 1. get_page() 함수 만들기 -> url을 입력받아서, dl파일을 return
# 2. page_to_items() 함수 만들기 -> dl을 입력받아, news_page list를 return
# 3. news_page로 df 만들기 news_df
# 4. df에서 csv 만들기 news_df.csv

import pandas as pd
import requests
from bs4 import BeautifulSoup

def get_page(url):
    """url 입력받아, dl 반환"""
    headers = {'User-Agent':'Mozilla/5.0'}
    page = url

    html = requests.get(page, headers=headers).text
    soup = BeautifulSoup(html, 'html.parser')
    type06 = soup.find('ul', {'class':'type06_headline'})

    dl = type06.find_all('dl')
    return dl

def page_to_items(dl):
    """dl 입력받아, news_page 리스트 반환"""
    news_page = []
    for item2 in dl:
        try:
            img = item2.find('dt', {'class':'photo'}).find('img')
            img_src = img['src']

            link = item2.find('dt', {'class':""}).find('a')
            link_href = link['href']

            link_text = link.text.replace('\t', '').replace('\r\n', '').strip().split('…')[0]

            content_text = item2.find('dd').text.replace('\t', '').replace('\n', '').split('…')[0]

            news_page.append([img_src, link_href, link_text, content_text])
        except:
            pass
    return news_page

get_page('https://news.naver.com/main/list.naver?mode=LSD&mid=sec&sid1=001')
news_page = page_to_items(dl)
news_page

# news_df = pd.DataFrame(news_page, columns=['img', 'link', 'title', 'news'])
# news_df.to_csv('news_df.csv')
# news_df
```
