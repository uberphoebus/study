```python
import pandas as pd
import numpy as np

list = [1, 2, 3, 3, 4, 4]
print(type(list))
print(list)

# list to srs
s = pd.Series(list); print(s)

print(pd.Series(list).value_counts())

# list to arr
print(np.array(list)) # np.array(list)
print(s.values)       # srs.values

print(type(np.array(list)))
print(type(s.values))

# arr는 변수로 만들 수 없고, 메서드를 거쳐야만 함
# python 내부적으로 사용하는 타입, 단독으로 존재할 수 없음

# arr to series
arr = np.array(list)

srs = pd.Series(arr)
print(srs, type(srs))

# array-like ; iterable, dict, scalar

# dict
list = [1, 2, 3, 3, 4, 4, 1]
val2 = [1, 2, 3, 3, 4, 4, 0]
dict = {'val':list, 'val2':val2}

print(dict)
print(dict.items())  # return tuple
print(dict.values()) # return iterable
print(dict.keys())

print(dict['val'])
print(np.array(dict['val']))

# dict to df
print(pd.DataFrame(dict))

# list to df
print(pd.DataFrame(list))

list1 = [1, 2, 3]
list2 = [10, 20, 30]

df1 = pd.DataFrame([list1, list2], index=['col1', 'col2']).T
print(df1)

# arr to df
arr_reshape1 = np.array(list1).reshape(-1, 1)
arr_reshape2 = np.array(list2).reshape(-1, 1)

df = pd.DataFrame(arr_reshape1, columns=['col1*'])
df['col2*'] = arr_reshape2
print(df)

# srs를 세로로 길게 세우려면 arr로 변X환
# data = srs를 제외하고는 matrix([[]]) 형태

# list
list = [1, 'kim', 30]
df = pd.DataFrame([list])
print(df)

# df + srs

# 열 추가
list = [[1, 'kim', 26], [2, 'hong', 20]]
df = pd.DataFrame(list, columns=['seq', 'name', 'age'])
df['addr'] = ['서울', '인천']

# 행 추가(레코드)
rec1 = [4, 'park', 40, '안양']
df.loc[2] = rec1

rec2 = [5, 'lee', 50, '강원']
# df.append( pd.Series(rec2, index=df.columns), ignore_index=True)

print(df)
```

```python
# mission 1 ; merge all
emp_all_df = pd.concat([emp_df, emp_new_df, emp_old_df], axis=0, ignore_index=True)
emp_all_df
```
