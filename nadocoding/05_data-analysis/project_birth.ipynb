```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('stat_142801.xls', skiprows=2, nrows=2, index_col=0).T
df.columns = ['출생아 수', '합계 출산율']
df
```

```python
fig, ax1 = plt.subplots(figsize=(13, 5))
fig.suptitle('출생아 수 및 합계출산율')
ax2 = ax1.twinx()

ax1.bar(df.index, df['출생아 수'], color='#ff812d')
ax2.plot(df.index, df['합계 출산율'], color='#ffd100', marker='o', ms=10, lw=4, mec='w', mew=3)

ax1.set_ylabel('출생아 수 (천 명)')
ax2.set_ylabel('합계 출산율 (가임여성 1명당 명')

ax1.set_ylim(250, 700)
ax1.set_yticks([300, 400, 500, 600])
ax2.set_ylim(0, 1.5)
ax2.set_yticks([0, 1])

for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 12, val, ha='center')
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.05, val, ha='center')

plt.show()
```
