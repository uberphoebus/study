```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df_m = pd.read_csv(
    '201108_201108_연령별인구현황_월간.csv', 
    encoding='cp949', index_col='행정구역', 
    usecols=[0] + [i for i in range(3, 24)]
)
df_m.columns = [col[11:] for col in df_m.columns]
df_m.index = [idx[:-14] for idx in df_m.index]
df_m.iloc[0] = df_m.iloc[0].str.replace(',', '').astype(int)

df_w = pd.read_csv(
    '201108_201108_연령별인구현황_월간.csv', 
    encoding='cp949', index_col='행정구역', 
    usecols=[0] + [i for i in range(26, 47)]
)
df_w.columns = [col[11:] for col in df_w.columns]
df_w.index = [idx[:-14] for idx in df_w.index]
df_w.iloc[0] = df_w.iloc[0].str.replace(',', '').astype(int)
```

```python
plt.figure(figsize=(10, 7))
plt.barh(df_m.columns, -df_m.iloc[0] // 1000)
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2011년 대한민국 인구 피라미드')
plt.savefig('2011년 대한민국 인구 피라미드.png', dpi=100)
```
