# 12. 데이터 수정

```python
temp = df.copy()
temp['학교'].replace({'북산고':'상북고', '능남고':'무슨고'}, inplace=True)
```

```python
temp['SW특기'] = temp['SW특기'].str.lower()
temp
```

```python
temp['학교'] = temp['학교'] + '등학교'
temp
```

```python
temp['총합'] = temp['국어'] + temp['영어'] + temp['수학'] + temp['사회'] + temp['과학']
temp
```
