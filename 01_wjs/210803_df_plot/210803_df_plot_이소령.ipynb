## 오차 막대를 사용한 폴로팅

```python
ix3 = pd.MultiIndex.from_arrays([["a", "a", "a", "a", "a", "b", "b", "b", "b", "b"],
                                ["foo", "foo", "foo", "bar", "bar", "foo", "foo", "bar", "bar", "bar"],],
                                names=["letter", "word"],)
ix3
```

```python
df3 = pd.DataFrame({"data1": [9, 3, 2, 4, 3, 2, 4, 6, 3, 2],
                    "data2": [9, 6, 5, 7, 5, 4, 5, 6, 5, 1],},
                     index=ix3,)
df3
```

```python
gp3 = df3.groupby(level=('letter', 'word'))
```

```python
means = gp3.mean()
errors = gp3.std()
```

```python
means
```

```python
errors
```

```python
fig, ax = plt.subplots()
means.plot.bar(yerr=errors, ax=ax, capsize=4, rot=0) #capsize: 오차막대 너비
```

```python
#비대칭 막대를 이용한 최소/최대 범위 표시
mins = gp3.min()
maxs = gp3.max()
errors = [[means[c] - mins[c], maxs[c] - mins[c]] for c in df3.columns]
fig, ax = plt.subplots()
means.plot.bar(yerr=errors, ax=ax, capsize=4, rot=0)
```
