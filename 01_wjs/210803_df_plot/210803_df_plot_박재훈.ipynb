# basic plotting: plot

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
print(ts)
ts.plot()
```

```python
# label 붙이기

df = pd.DataFrame(np.random.randn(1000, 4),
                  index=ts.index,
                  columns=list('ABCD')
                 )
df = df.cumsum()
print(df)

plt.figure()
df.plot()
```

```python
# 컬럼의 값을 축에 설정

df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
print(df3)

df3['A'] = pd.Series(list(range(len(df))))

print(df3)

df3.plot(x='A', y='B') # x축 A, y축 B
```

## box plots; df.plot.box(), df.boxplot()
distribution of values within each column

```python
# sample
df = pd.DataFrame(np.random.rand(10, 6),
                  columns=['A', 'B', 'C', 'D', 'E', 'F']
                 )
df.plot.box()
```

```python
# color = {'boxes':, 'whiskers':, 'medians':, 'caps': }
color = {'boxes':'DarkGreen', # 박스 색
         'whiskers':'DarkOrange', # 기둥 색
         'medians':'Darkblue', # 중앙값 색
         'caps':'Gray'} # 끝 색
df.plot.box(color=color,
            sym='r+') # 이상값의 색/형태
```

```python
# 수평과 위치; vert=False, positions=[]
df.plot.box(vert=False, positions=[1, 5, 7, 9, 2, 3])
```

```python
# 컬럼 기준 그룹화; by=column
df = pd.DataFrame(np.random.rand(10, 2), columns=['Col1', 'Col2'])
df['X'] = pd.Series(['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'])
print(df)

plt.figure()
bp = df.boxplot(by='X')
```

```python
# multiple column grouping; by=[]
df = pd.DataFrame(np.random.rand(10, 3), columns=['Col1', 'Col2', 'Col3'])
df['X'] = pd.Series(['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'])
df['Y'] = pd.Series(['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'])
print(df)

plt.figure()
bp = df.boxplot(column=['Col1', 'Col2'], by=['X', 'Y'])
```

```python
# grouping; df.groupby(cl).boxplot()
np.random.seed(1234)
df_box = pd.DataFrame(np.random.randn(50, 2))
df_box['g'] = np.random.choice(['A', 'B'], size=50)
print(df_box)

bp = df_box.groupby('g').boxplot()
```

## plotting with error bars
xerr, yerr

```python
# grouping means, standard deviations
ix3 = pd.MultiIndex.from_arrays([['a', 'a', 'a', 'a', 'a', 'b', 'b', 'b', 'b', 'b'],
                                 ['foo', 'foo', 'foo', 'bar', 'bar', 'foo', 'foo', 'bar', 'bar', 'bar'],
                                ],
                                names=['letter', 'word']
                               )
print(ix3)

df3 = pd.DataFrame({'data1':[9, 3, 2, 4, 3, 2, 4, 6, 3, 2],
                    'data2':[9, 6, 5, 7, 5, 4, 5, 6, 5, 1]
                   },
                   index=ix3
                  )
print(df3)

gp3 = df3.groupby(level=('letter', 'word'))
means = gp3.mean()
errors = gp3.std()
print(means), print(errors)

fig, ax = plt.subplots()
means.plot.bar(yerr=errors, ax=ax, capsize=4, rot=0)
```

```python
# min/max range
mins = gp3.min()
maxs = gp3.max()

errors = [[means[c] - mins[c], maxs[c] - means[c]] for c in df3.columns]
print(errors)

fig, ax = plt.subplots()
means.plot.bar(yerr=errors, ax=ax, capsize=4, rot=0)
```

# plotting directly with matplotlib

```python
price = pd.Series(np.random.randn(150).cumsum(),
                  index=pd.date_range('2000-1-1', periods=150, freq='B')
                 )
print(price)
ma = price.rolling(20).mean()
mstd = price.rolling(20).std()
print(ma)
print(mstd)

plt.figure()
plt.plot(price.index, price, 'k')
plt.plot(ma.index, ma, 'b')
plt.fill_between(mstd.index, ma - 2 * mstd, ma + 2 * mstd, color='b', alpha=0.2)
```
