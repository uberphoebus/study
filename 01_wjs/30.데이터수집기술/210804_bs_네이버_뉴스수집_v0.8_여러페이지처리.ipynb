## ( 실습 ) 
1. 네이버 뉴스 속보 수집
   - 날짜, page 구분 
   - 날짜 . 8월3일, 8월4일 데이터 
   - 날짜별 page = 3page (1,2,3)
2. get_page() 함수 만들기
   - input ->  url을 입력받아서
   - output -> dl
3. page_to_items() 함수 만들기
   - input -> dl 
   - output -> news_page list를 Return 
4. news_page로 DataFrame 만들기 -> news_df
   - ["날짜","이미지링크","기사링크","기사제목","기사내용"]
5. DataFrame에서 csv 만들기 -> news_df.csv 

```python
## 네이버 뉴스 속보를 수집 
import requests 
from bs4 import BeautifulSoup 
headers = { "User-Agent":"Mozilla/5.0"}

def get_page(url):    
    html = requests.get(url, headers=headers)
    soup = BeautifulSoup(html.text)  # "html.parser"  
    type06 = soup.find("ul",{"class":"type06_headline"})
    dl = type06.find_all("dl")
    return dl

def page_to_items(dl):
    news_page = []

    for item2 in dl:
        try:
            # 기사 이미지 
            img = item2.find("dt",{"class":"photo"}).find("img")
            img_src = img["src"]      
            # 기사 링크, 타이틀 
            link = item2.find("dt", {"class":""}).find("a")
            link_href = link["href"]        
            link_text =link.text.replace("\t","").replace("\r\n","").strip()     
            # 기사 내용 
            content = item2.find("dd")
            content_text = content.text.replace("\n","").split("…")[0] 

            news_page.append([img_src, link_href, link_text, content_text])

        except: 
            print("No Image or No Contents!! ")
    return news_page
```

```python
import pandas as pd
import time 

base_url = "https://news.naver.com/main/list.nhn?mode=LSD&mid=sec&sid1=001"
date = ["20210803","20210804"]
page = [1,2,3]

init = 0

for i in date:
    for j in page:
        url = base_url + "&date=" + i + "&page=" + str(j)
        print(url)
        
        time.sleep(0.5)
        get_dl = get_page(url)
        news_page = page_to_items(get_dl)        

        news_page_df = pd.DataFrame(news_page, 
                      columns = ["이미지링크","기사링크","기사제목","기사내용"])
        news_page_df["날짜"] = i
        news_page_df["page"] = j 
        
        if init ==0 : 
            total_df = news_page_df.copy()
            init = 1
        else:
            total_df = pd.concat([total_df, news_page_df], ignore_index=True)
        
        print("total_df = ", total_df.shape)
        
print("완료 = ", total_df.shape)        
    
# 컬럼 순서 바꾸기
total_df = total_df[["날짜","page","기사제목","기사내용","이미지링크","기사링크"]]

# csv 화일 만들기
total_df.to_csv("c:/jswoo/naver_news_page.csv", encoding = "utf-8-sig", index=False)
pd.read_csv("c:/jswoo/naver_news_page.csv")
```
