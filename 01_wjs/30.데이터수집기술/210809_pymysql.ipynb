## ( 실습. pymysql 활용 Table 만들고 데이터 넣기 ) 
1. member 테이블 생성 (pydb에) 
   - mem_no  int(3)   primary key 
   - name    varchar(30)
   - age     int(3)
   - height  int(3)
2. 데이터 입력 ( insert문 sql로 만들어서 한건씩 실행 ) 
   - 1, 홍길동, 35, 170
   - 2, 이순신, 65, 190
   - 3, 유관순, 18, 160
   - 4, 우수한, 55, 155
3. cursor.execute(sql) 4번 실행

```python
import pymysql
db = pymysql.connect(host="localhost",port=3306,
                     user="root",passwd="pass",
                     db="pydb", charset="utf8")
cursor = db.cursor()
## 테이블 생성 
sql = """
      create table if not exists member (
             mem_no    int(3),
             name    varchar(50),
             age     int(3),
             height  int(3), 
             primary key(mem_no)); 
"""
#print(sql)
cursor.execute(sql)
```

```python
sql = "insert into member value (1, '홍길동', 35, 170)"
#print(sql)
res = cursor.execute(sql) 
#print("res1 = ", res)

sql = "insert into member value (2, '이순신', 65, 190)"
#print(sql)
res = cursor.execute(sql) 
#print("res2 = ", res)

sql = "insert into member value (3, '유관순', 18, 160)"
#print(sql)
res = cursor.execute(sql) 
#print("res3 = ", res)

sql = "insert into member value (4, '우수한', 55, 155)"
#print(sql)
res = cursor.execute(sql) 
#print("res4 = ", res)

db.commit()
```

```python
sql = "insert into member value (5, '우수한', 55, 155)"
res = cursor.execute(sql) 

db.commit()
```

##  (실습.. )
1. mem_no가 4인 레코드의 name을  강감찬으로 변경하세요. 

```python
# UPDATE  테이블명
#    set  컬럼명 = "변경자료"
#  where 조건문 
```

```python
sql = """
      update member 
        set  name = "강감찬"
      where  mem_no = 4
      
"""

cursor.execute(sql)
db.commit() 
```
