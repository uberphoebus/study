```python
N = 5
menMeans = (20, 35, 30, 35, -27)
womenMeans = (25, 32, 34, 20, -25)
menStd = (2, 3, 4, 1, 2)
womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence
```

```python
fig, ax = plt.subplots()

p1 = ax.bar(ind, menMeans, width, yerr=menStd, label='Men')
p2 = ax.bar(ind, womenMeans, width,
            bottom=menMeans, yerr=womenStd, label='Women')

ax.axhline(0, color='grey', linewidth=0.8)
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()

# Label with label_type 'center' instead of the default 'edge'
ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2)

plt.show()
```

## ( 함수 만들어 그리기 )

```python
def my_plotter(ax, data1, data2, param_dict):
    out = ax.plot(data1, data2, **param_dict)
    return out 
```

```python
import warnings
warnings.filterwarnings(action="ignore")

data1, data2, data3, data4 = np.random.randn(4,100)
# data1 = np.linspace(0, 2, 100)
# data2 = np.linspace(0, 2, 100)
# data3 = np.linspace(0, 2, 100)
# data4 = np.linspace(0, 2, 100)

fix, ax = plt.subplots(1,1)
my_plotter(ax,data1,data2, {'marker':'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {"marker":"x"})
my_plotter(ax2, data3, data4, {"marker":"o"})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {"marker":"x", "color":"red"})
my_plotter(ax2, data3, data4, {"marker":"o", "color":"green"})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, {"marker":"x", "color":"red", "linewidth":3})
my_plotter(ax2, data3, data4, {"marker":"o", "color":"green", "linewidth":20})
```

```python
fig, ax = plt.subplots(2,2)
my_plotter(ax[0][0], data1, data2, {"marker":"x", "color":"red"})
my_plotter(ax[0][1], data3, data4, {"marker":"o", "color":"green"})
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.pie(sizes, # 값
        labels=labels, # 분류, 구분, class
        autopct='%1.1f%%', # 값 표시, 
        explode=(0.3, 0, 0, 0),
        shadow=True,
        startangle= 180 )  
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
