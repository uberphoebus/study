### (실습..  학생별 총점과 평균은 ? )

```python
# 1. 데이터 확인하기
for i in class_a.items():
    print(i)
```

```python
# 2. 점수 확인하기 
for i in class_a.items():
    print(i[0], " ** ", i[1])
```

```python
# 3. 개인별 점수 더하기 
for i in class_a.items():
    print(i[0], " ** ", i[1][0]+i[1][1]+i[1][2]+i[1][3])
```

```python
# 3. 개인별 점수 더하기 
for i in class_a.items():
    tot = 0
    for score in i[1]:
        tot = tot + score
    print(i[0], " ** ", tot)
```

```python
# 4. 개인별 점수 합계, 평균
for i in class_a.items():
    tot = 0
    for score in i[1]:
        tot = tot + score
    print(i[0], " 총점=", tot, " 평균=", tot/len(i[1]))
```

```python
# 5. 개인별 점수 합계, 평균 저장하기 (dict 방식)
tot_dict = {}
for i in class_a.items():
    tot = 0
    for score in i[1]:
        tot = tot + score
    tot_dict[i[0]] = [tot, tot/len(i[1])]
print(tot_dict)
```

```python
# 6. 개인별 점수 합계, 평균 저장하기 (list 방식)
tot_list = []
for i in class_a.items():
    tot = 0
    for score in i[1]:
        tot = tot + score
    tot_list.append([i[0],tot,tot/len(i[1])])
print(tot_list)
```
