### 6) find
    : 위치 알려주기
    
* 문자열 중 문자 o가 처음 나온 위치를 알려준다. 
* 만약 찾는 문자나 문자열이 존재하지 않는다면 -1을 반환한다.

```python
c = "Today is Monday"

# 처음 나온 위치를 알려줌
c.find('o')  
```

```python
# 찾는 문자나 문자열이 존재하지 않음
c.find('z')   
```

### 7) index

* find 함수와 다른 점: **문자열 안에 존재하지 않는 문자를 찾으면 오류가 발생**한다.

```python
c = "Today is Monday"

# 처음 나온 위치를 알려줌
c.index('o')
```

```python
# 찾는 문자나 문자열이 존재하지 않음 => 오류 발생
c.index('z')   
```

### 15) split
    : 문자열 나누기
* split()

 : 괄호 안에 아무 값도 넣어 주지 않으면 **공백(스페이스, 탭, 엔터 등)**을 기준으로 문자열을 나누어 준다.
 
 
* split(',')

 : 괄호 안에 특정 값이 있을 경우에는 **괄호 안의 값을 구분자**로 해서 문자열을 나누어 준다. 

```python
# 공백(스페이스, 탭, 엔터 등) 기준
k = "Python is easy."
k.split()
```

```python
# 괄호 안 값의 구분자 기준
l = "Python,is,easy."
l.split(',')
```

### 8) pop
    : 리스트 요소 끄집어내기
* pop()

: 리스트의 맨 마지막 요소를 돌려주고 그 요소는 삭제한다.


* pop(x) 

:리스트의 x번째 요소를 돌려주고 그 요소는 삭제한다.

```python
# pop(): 맨 마지막 요소 돌려주고 그 요소는 삭제
h = [1,2,3]
h.pop()
```

```python
# pop(1): 1번째 요소를 돌려주고 
h = [1,2,3]
h.pop(1)
```

```python
# 그 요소는 삭제(1번째 요소 = 2)
h
```

### 9) reverse
    : 리스트 뒤집기 
* reverse 함수는 리스트를 역순으로 뒤집어 준다. 
* 이때 리스트 요소들을 순서대로 정렬한 다음 다시 역순으로 정렬하는 것이 아니라 **현재의 리스트를 그대로 거꾸로 뒤집는다**.

```python
# 현재의 리스트 그대로 거꾸로 뒤집기
i = ['a', 'c', 'b']
i.reverse() ;i
```

### 7) get
    : Key로 Value얻기
* get(x) 함수는 x라는 Key에 대응되는 Value를 돌려준다. 
* a.get('name')은 a['name']을 사용했을 때와 동일한 결과값을 돌려받는다.
* 존재하지 않는 키로 값을 가져오려고 할 경우,
  Key 오류를 발생시키고 a.get('')는 None(거짓)을 돌려준다.
* 딕셔너리 안에 찾으려는 Key 값이 없을 경우 미리 정해 둔 디폴트 값을 대신 가져오게 하고 싶을 때에는 get(x, '디폴트 값')을 사용하면 편리하다.

```python
a = {'name': 'Joy', 'age': '17', 'sex': 'girl'}
a.get('name') 
```

```python
# 존재하지 않는 키로 값을 가져오려고 할 경우

print(a.get('nokey'))  # None(거짓)
```

```python
print(a['nokey'])      # 오류
```

```python
# get(x, '디폴트 값')

a.get('birth', '1225')  
# a 딕셔너리에는 'birth'에 해당하는 값 없음. 따라서 디폴트 값인 '1225'를 돌려줌.
```

### 8) in
    : 해당 Key가 딕셔너리 안에 있는지 조사

```python
# 'name' 문자열은 a 딕셔너리의 Key 중 하나
'name' in a
```

```python
# 'birth'은 a 딕셔너리 안에 존재하지 않는 Key
'birth' in a
```

### 9) update
    : 기존 dictionary의 데이터를 한꺼번에 바꿀때 유용

```python
a = {'name': 'Joy', 'age': '17', 'sex': 'girl'}
a.update({'birth': '1225'}) ;a
```
