##(String Methods)

```python
txt = 'hello, and welcome to my world.'
```

```python
#문자열이 지정된 값으로 끝나는지 확인
txt.endswith('.')
```

##(List Methods)

```python
#List에 요소 추가
fru = ['apple', 'banana', 'cherry']
fru.append('orange')
```

```python
print(fru)
```

```python
#List 모든 요소 제거
fru.clear()
print(fru)
```

```python
#복사본
fru1 = ['apple', 'banana', 'melon']
fru1.copy()
```

```python
#지정된 값을 가진 요소의 수 반환
fru1.count('apple')
```

```python
#List.extend(List1) List에 List1 추가
fru2 = ['orange', 'mango']
fru1.extend(fru2)
```

```python
fru1
```

```python
#지정된 값의 위치
fru1.index('banana')
```

```python
#지정된 위치에 지정된 값 추가
fru1.insert(1, 'peach')
```

```python
print(fru1)
```

```python
#지정된 위치의 요소 제거
fru1.pop(1)
print(fru1)
```

```python
fru1 = ['apple', 'peach', 'banana', 'melon', 'orange', 'mango']
```

```python
#지정된 요소 제거
fru1.remove('banana')
```

```python
#List 순서 반대
fru1.reverse()
```

```python
#오름차순으로 정렬
fru1.sort()
```

##(Dictionary Method)

```python
car = {'brand' : 'Ford', 'model' : 'Mustang', 'year' : 1964}
```

```python
#모든 요소 제거
car.clear()
```

```python
car
```

```python
#복사본
per = {'name': 'soryeong', 'age':'25'}
z = per.copy()
```

```python
print(z)
```

```python
#지정된 키와 지정된 값이 있는 사전 반환
x = ('key1', 'key2', 'key3')
y = 0
xy = dict.fromkeys(x, y)
print(xy)
```

```python
#지정된 키를 통해 값 반환
per.get('name')
```

```python
car1 = {'brand' : 'Ford', 'model' : 'Mustang', 'year' : 1964}
```

```python
car1.items() #for문에 사용
```

```python
car1.keys()
```

```python
car1.values()
```

```python
#지정된 항목 제거 dict.pop(key, defaultvalue)
car1.pop('model')
```

```python
#마지막 항목 제거
car1.popitem()
```
